```python
import os
import glob
import re
import numpy as np
import pandas as pd
import rasterio
```

```python
raster_folder = "/Users/cherryleheu/Documents/HCDP/climate_report/dependencies/rainfall"

records = []

for tif in sorted(glob.glob(os.path.join(raster_folder, "rainfall_*_06.tif"))):
    year = int(re.search(r"rainfall_(\d{4})_06\.tif", os.path.basename(tif)).group(1))

    with rasterio.open(tif) as src:
        arr = src.read(1, masked=True)

    mean_mm = arr.mean()
    mean_in = mean_mm / 25.4

    records.append({
        "year": year,
        "mean_rainfall_in": float(mean_in),
    })

df = pd.DataFrame(records).sort_values("year").reset_index(drop=True)
df
```

```python
# rank 1 = wettest June (highest statewide total rainfall)
df["rank"] = df["mean_rainfall_in"].rank(method="min", ascending=False).astype(int)
df = df.sort_values("rank").reset_index(drop=True)
df
```
