```python
import requests
from dotenv import load_dotenv
import os

load_dotenv('../envs/dev.env')

token = os.getenv("HCDP_API_TOKEN")

headers = {
    "Authorization": f"Bearer {token}"
}


body = {
    "email": "cherryle@hawaii.edu",
    "county": ["honolulu"],
    "moku": ["wahiawa"]
}
```

```python
url = "https://api.hcdp.ikewai.org/mesonet/climate_report/subscribe"
res = requests.post(url, json = body, headers = headers)
res.raise_for_status()
print("User subscribed")
```

```python
#Get existing subscription info. If new subscription, userID is None. 
params = {
  "email": "cherryle@hawaii.edu"
}
url = "https://api.hcdp.ikewai.org/mesonet/climate_report/email_lookup"
res = requests.get(url, json = body, params = params, headers = headers)
res.raise_for_status()
data = res.json()
userID = data["userID"]

url = f"https://api.hcdp.ikewai.org/mesonet/climate_report/subscription/{userID}"
res = requests.get(url, headers = headers)
json = res.json()

new_body = {
    "email": "cherryle@hawaii.edu",
    "county": ["hawaii"],
    "moku": ["kohala"]
}

LIST_FIELDS = ["county", "moku", "ahupuaa", "watershed"]

updated_body = {"email": json["email"]}

for f in LIST_FIELDS:
    old_vals = json.get(f, []) 
    new_vals = new_body.get(f, []) 
    updated_body[f] = list(dict.fromkeys(old_vals + new_vals))

updated_body
```

```python
url = f"https://api.hcdp.ikewai.org/mesonet/climate_report/subscription/{userID}"
res = requests.patch(url, json = updated_body, headers = headers)
res.raise_for_status()
print("Subscription updated")
```
