```python
import numpy as np
import rasterio

def tif_value_at_xy(tif_path, x, y, band=1):
    """
    Return the raster value at coordinate (x, y) in the GeoTIFF's CRS.
    If outside bounds or nodata, returns None.
    """
    with rasterio.open(tif_path) as src:
        # Convert map (x,y) -> (row,col)
        row, col = src.index(x, y)

        # Bounds check
        if row < 0 or row >= src.height or col < 0 or col >= src.width:
            return None

        # Read just that one pixel
        val = src.read(band, window=((row, row + 1), (col, col + 1)))[0, 0]

        # Nodata handling
        nodata = src.nodata
        if nodata is not None:
            # works for ints/floats; use isclose for float nodata if you like
            if np.isnan(nodata):
                if np.isnan(val):
                    return None
            else:
                if val == nodata:
                    return None

        return float(val) if np.issubdtype(type(val), np.number) else val

# Example
tif_path = "/Users/cherryleheu/Documents/HCDP/Data/annual/rainfall/calendar_year/annual_rainfall_2025_agg.tif"
x, y = -155.29129,19.9321
mm = tif_value_at_xy(tif_path, x, y)
mm/25.4
```
