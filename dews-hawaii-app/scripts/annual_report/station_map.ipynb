```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

df = pd.read_csv("station_metadata.csv")

# adjust these if your column names differ
LAT_COL = "lat"
LON_COL = "lng"
STATUS_COL = "status"   # values like: active, planned, inactive

ACTIVE = "#0072B2"   # blue
PLANNED = "#D55E00"

# keep only active + planned
df = df[df[STATUS_COL].str.lower().isin(["active", "planned"])].copy()

# points -> GeoDataFrame
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df[LON_COL], df[LAT_COL]),
    crs="EPSG:4326",
).to_crs("EPSG:3857")

# plot by status (different colors)
fig, ax = plt.subplots(figsize=(10, 8))
gdf[gdf[STATUS_COL].str.lower() == "active"].plot(
    ax=ax, markersize=35, color=ACTIVE, label="Active",
    edgecolor="white", linewidth=0.8, alpha=0.95
)

gdf[gdf[STATUS_COL].str.lower() == "planned"].plot(
    ax=ax, markersize=35, color=PLANNED, label="Planned",
    edgecolor="white", linewidth=0.8, alpha=0.95
)
ctx.add_basemap(ax, source=ctx.providers.Esri.WorldPhysical)
ax.legend()
ax.set_axis_off()
plt.savefig('../../public/climate-summary/pubs/station_map.png', dpi=600, bbox_inches='tight', pad_inches=0.1, transparent=True)
plt.show()

```
