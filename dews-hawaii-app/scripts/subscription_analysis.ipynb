# Climate Report Subscription Analysis

```python
import os
import requests
from collections import Counter
from dotenv import load_dotenv

load_dotenv('../envs/dev.env')
api_token = os.environ.get("HCDP_ADMIN_TOKEN")
headers = {"Authorization": f"Bearer {api_token}"}

url = "https://api.hcdp.ikewai.org/mesonet/climate_report/subscriptions"
res = requests.get(url, headers=headers)
res.raise_for_status()
subscriptions = res.json()

print(f"Total subscriptions: {len(subscriptions)}")
```

## Category Breakdown

```python
total = len(subscriptions)

statewide_only = 0
has_island = 0
has_ahupuaa = 0
has_watershed = 0
has_moku = 0
has_climate = 0

island_counter = Counter()
ahupuaa_counter = Counter()
watershed_counter = Counter()
moku_counter = Counter()
climate_counter = Counter()

for sub in subscriptions:
    islands   = sub.get("island", [])
    ahupuaa   = sub.get("ahupuaa", [])
    watershed = sub.get("watershed", [])
    moku      = sub.get("moku", [])
    climate   = sub.get("climate", [])

    if islands == ["Statewide"] and not ahupuaa and not watershed and not moku and not climate:
        statewide_only += 1

    if islands:   has_island += 1
    if ahupuaa:   has_ahupuaa += 1
    if watershed: has_watershed += 1
    if moku:      has_moku += 1
    if climate:   has_climate += 1

    for i in islands:   island_counter[i] += 1
    for a in ahupuaa:   ahupuaa_counter[a] += 1
    for w in watershed: watershed_counter[w] += 1
    for m in moku:      moku_counter[m] += 1
    for c in climate:   climate_counter[c] += 1

print(f"{'Category':<30} {'Count':>6}  {'%':>6}")
print("-" * 46)
print(f"{'Statewide only':<30} {statewide_only:>6}  {statewide_only/total*100:>5.1f}%")
print(f"{'Has island selection':<30} {has_island:>6}  {has_island/total*100:>5.1f}%")
print(f"{'Has ahupuaa selection':<30} {has_ahupuaa:>6}  {has_ahupuaa/total*100:>5.1f}%")
print(f"{'Has watershed selection':<30} {has_watershed:>6}  {has_watershed/total*100:>5.1f}%")
print(f"{'Has moku selection':<30} {has_moku:>6}  {has_moku/total*100:>5.1f}%")
print(f"{'Has climate selection':<30} {has_climate:>6}  {has_climate/total*100:>5.1f}%")
```

## Island Selections

```python
print(f"{'Island':<25} {'Count':>6}")
print("-" * 33)
for name, count in island_counter.most_common():
    print(f"{name:<25} {count:>6}")
```

## Detailed Selections (Ahupuaa, Watershed, Moku, Climate)

```python
def print_counter(label, counter, top=20):
    if not counter:
        print(f"No {label} selections.\n")
        return
    print(f"{'— ' + label + ' —'}")
    print(f"{'Name':<40} {'Count':>6}")
    print("-" * 48)
    for name, count in counter.most_common(top):
        print(f"{name:<40} {count:>6}")
    print()

print_counter("Ahupuaa", ahupuaa_counter)
print_counter("Watershed", watershed_counter)
print_counter("Moku", moku_counter)
print_counter("Climate", climate_counter)
```
