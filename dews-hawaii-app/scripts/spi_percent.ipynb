```python
import numpy as np
import pandas as pd
import xarray as xr
import xclim.indices as xci

n_days = 10          # number of days
ny, nx = 4, 5        # grid size (eg lat lon)
times = pd.date_range("2025-01-01", periods=n_days, freq="D") #1/1/25 to 1/10/25

# --- Dummy Data ---
# Total daily rainfall (mm)
pr_data = np.random.uniform(0, 50, size=(n_days, ny, nx))  # e.g., 0–50 mm/day total

# Daily max temperature (°C)
tasmax_data = np.random.uniform(20, 35, size=(n_days, ny, nx))

# Mean annual precipitation climatology (mm/year)
pr_annual_data = np.random.uniform(1000, 3000, size=(ny, nx))

# Initial KBDI (mm)
kbdi0_data = np.zeros((ny, nx)) #Starting kbdi will be 0. I assume the prior kbdi will automatically change (ie KBDI for day 2 will automatically use day 1's KBDI)

# --- Convert to DataArrays ---
pr = xr.DataArray(
    pr_data,
    dims=("time", "lat", "lon"),
    coords={"time": times, "lat": np.arange(ny), "lon": np.arange(nx)},
    name="pr",
    attrs={"units": "mm/day"},  # make sure this is explicitly in mm/day for it to work
)

tasmax = xr.DataArray(
    tasmax_data,
    dims=("time", "lat", "lon"),
    coords={"time": times, "lat": np.arange(ny), "lon": np.arange(nx)},
    name="tasmax",
    attrs={"units": "degC"},
)

pr_annual = xr.DataArray(
    pr_annual_data,
    dims=("lat", "lon"),
    coords={"lat": np.arange(ny), "lon": np.arange(nx)},
    name="pr_annual",
    attrs={"units": "mm/year"},
)

kbdi0 = xr.DataArray(
    kbdi0_data,
    dims=("lat", "lon"),
    coords={"lat": np.arange(ny), "lon": np.arange(nx)},
    name="kbdi0",
    attrs={"units": "mm/day"},
)

# --- Compute KBDI using package---
kbdi = xci.keetch_byram_drought_index(pr, tasmax, pr_annual, kbdi0=kbdi0)
```
