```python
#Save PNG of annual rainfall aggregated for website display
import rasterio
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib as mpl

date = 15
out_png = f"../public/storm_site/2026_03_{date:02d}.png"

with rasterio.open(f"./storm_data/rainfall_2026-03-{date:02d}.tif") as src:
    data = src.read(1).astype(float)
    nodata = src.nodata
    bounds = src.bounds  

masked = np.ma.masked_where(data == nodata, data) / 25.4

vmax = np.nanmax(masked)

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(
    masked,
    cmap = 'viridis_r',
    extent=(bounds.left, bounds.right, bounds.bottom, bounds.top),
    vmin=0,
    vmax=vmax,
)
print(f"vmax: {vmax}")
ax.set_axis_off()  # removes ticks + border/frame
# cbar = fig.colorbar(im, ax=ax)
fig.savefig(out_png, dpi=300, bbox_inches="tight", pad_inches=0.1, transparent=True)
plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

date = 15

with rasterio.open(f"./storm_data/2026-03-10_to_2026-03-{date:02d}.tif") as src:
    data = src.read(1).astype(float)
    nodata = src.nodata
    bounds = src.bounds

masked = np.ma.masked_where(data == nodata, data) / 25.4

bin_edges = [0, 0.1, 0.5, 1, 2, 4, 8, 12, 20, 30, 40, 52]

colors = [
    "#f7fcf5",
    "#e5f5e0",
    "#c7e9c0",
    "#a1d99b",
    "#74c476",
    "#31a354",
    "#fdd835",
    "#fb8c00",
    "#e53935",
    "#8e24aa",
    "#4a148c"
]

cmap = mcolors.ListedColormap(colors)
norm = mcolors.BoundaryNorm(bin_edges, cmap.N, clip=True)

fig, ax = plt.subplots(figsize=(10, 8))

im = ax.imshow(
    masked,
    cmap=cmap,
    norm=norm,
    extent=(bounds.left, bounds.right, bounds.bottom, bounds.top)
)

cbar = fig.colorbar(
    im,
    ax=ax,
    boundaries=bin_edges,
    ticks=bin_edges[:-1]
)

cbar.set_label("Rainfall (inches)", fontsize=12)

cbar.ax.set_yticklabels([
    "<0.1",
    "0.1–0.5",
    "0.5–1",
    "1–2",
    "2–4",
    "4–8",
    "8–12",
    "12–20",
    "20–30",
    "30–40",
    "40+"
])

plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm

date = 15

with rasterio.open(f"./storm_data/2026-03-10_to_2026-03-{date:02d}.tif") as src:
    data = src.read(1).astype(float)
    nodata = src.nodata
    bounds = src.bounds

masked = np.ma.masked_where(data == nodata, data) / 25.4

base = cm.get_cmap("gist_rainbow")
colors = base(np.linspace(0, 0.8, 100))
cmap = mcolors.ListedColormap(colors)
cmap.set_bad((0, 0, 0, 0))

norm = mcolors.SymLogNorm(
    linthresh=0.4,
    linscale=1,
    vmin=0,
    vmax=51,
    base=10
)

fig, ax = plt.subplots(figsize=(10, 8))

im = ax.imshow(
    masked,
    cmap=cmap,
    norm=norm,
    extent=(bounds.left, bounds.right, bounds.bottom, bounds.top)
)

ticks = [0, 0.4, 1, 2, 4, 8, 16, 32, 51]

cbar = fig.colorbar(im, ax=ax, ticks=ticks)
cbar.set_label("Rainfall (inches)", fontsize=12)
cbar.ax.set_yticklabels([f"{t:g} in" for t in ticks])

plt.show()
```
